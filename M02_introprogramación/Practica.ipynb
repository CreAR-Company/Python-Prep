```python
import pandas as pd
import statsmodels.api as sm

# Supongamos que tienes un DataFrame llamado 'datos'
# con columnas 'ingresos' y 'calificación_compra'

# Calcula el promedio actual de calificación de compra
promedio_actual = datos['calificación_compra'].mean()

# Aumenta el promedio en un 20 %
promedio_aumentado = promedio_actual * 1.2

# Crea un nuevo DataFrame con el promedio aumentado
datos_aumentados = datos.dropna().copy()
datos_aumentados['calificación_compra'] = promedio_aumentado

# Realiza una regresión lineal para predecir los ingresos
X = datos_aumentados['calificación_compra']
y = datos_aumentados['gastos']
X = sm.add_constant(X)  # Agrega una columna constante para el intercepto

model = sm.OLS(y, X).fit()

# Imprime los resultados de la regresión
print(model.summary())
```
